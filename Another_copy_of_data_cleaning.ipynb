```python

#@import dataset using pandas library
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/Customer Call List.xlsx')
df


```

```python
df =df.drop_duplicates()
df
```

```python

df["Last_Name"] =df["Last_Name"].str.strip("123._/")
df
```

```python
df =df.drop(columns="Not_Useful_Column")
df
```

```python
# prompt: Format all the phone numbers in this format 123-545-5421	 3digits-3 digits-4dig

df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10] if x else '')
df
```

```python
#@ remove Address Column into street_Address, state and Zip code
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', expand=True)
df = df.drop(columns="Address")
df

```

```python
#@ In the Paying Customer Column format the yes no into Y and N only
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
df
```

```python
#@ dont contact column do y and n
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df
```

```python
#@ remove Na and NaN with empty string for both Paying Customer	Do_Not_Contact
df = df.replace('N/a', '')
df = df.fillna('')
df

```

```python
#@ in the Do_Not_Contact coln drop the row that are Y
for x in df.index:
  if df.loc[x, "Do_Not_Contact"] == 'Y':
    df.drop(x, inplace=True)
df
```

```python
#@ remove the ones with no phone number
for x in df.index:
  if df.loc[x, "Phone_Number"] == '--':
    df.drop(x, inplace=True)
df
```

```python
#@ reset the SN
df = df.reset_index(drop=True)
df
```
